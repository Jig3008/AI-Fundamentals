Logical Conditions

```python
porridge = np.array([79, 65, 50, 63, 56, 90, 85, 98, 79, 51])

'''
Today we’re visiting the Goldilocks Porridge Festival,
sampling a selection of breakfast cereals and judging them based on their temperature (listed in Fahrenheit).
Create a logical condition that selects samples in the porridge array that are
less than 60, and save them to a variable named cold.
'''

cold = porridge[porridge < 60]
cold
```

```python
'''
Create a logical condition that finds all the samples that are higher than 80
and save them to a variable named hot.
'''

hot = porridge[porridge > 80]
hot
```

```python
'''
Create a logical condition that finds all the samples that are between 60 and
80 (inclusive) and save them to a variable named just_right.
'''

just_right = porridge[(porridge > 60) & (porridge <= 80)]
just_right
```

# Some percentiles have specific names:

1. The 25th percentile is called the first quartile
2. The 50th percentile is called the median
3. The 75th percentile is called the third quartile

The difference between the first and third quartile is a value called the **interquartile range [IQR].**

The interquartile range gives us an idea of how spread out our data is. The smaller the interquartile range value, the less variance in our dataset. The greater the value, the larger the variance.

```python
movies_watched = np.array([2, 3, 8, 0, 2, 4, 3, 1, 1, 0, 5, 1, 1, 7, 2])

first_quarter = np.percentile(movies_watched,25)
third_quarter = np.percentile(movies_watched,75)

interquartile_range = third_quarter - first_quarter

print(f'First Quarter is {first_quarter}, Third Quarter is {third_quarter} and the IQR is {interquartile_range}')
```

While the mean and median can tell us about the center of our data, they do not reflect the range of the data. That’s where standard deviation comes in.

Similar to the interquartile range, the **standard deviation tells us the spread of the data**. The larger the standard deviation, the more spread out our data is from the center. The smaller the standard deviation, the more the data is clustered around the mean.

```python
#Sample Practice Exercise

rainfall = np.array([5.21, 3.76, 3.27, 2.35, 1.89, 1.55, 0.65, 1.06, 1.72, 3.35, 4.82, 5.11])

rain_mean = np.mean(rainfall)

rain_median = np.median(rainfall)

first_quarter = np.percentile(rainfall,25)
third_quarter = np.percentile(rainfall,75)

interquartile_range = third_quarter - first_quarter

rain_std = np.std(rainfall)

print(f'The mean value is {rain_mean}. The median value is {rain_median}. The first quarter is {first_quarter}. The Third Quarter is {third_quarter}. The IQR is {interquartile_range}. The standard deviation is {rain_std}')
```
